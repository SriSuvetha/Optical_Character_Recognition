```python
import easyocr
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

**READING IMAGES**

```python
path= "/content/img.jpg"
read = easyocr.Reader(['en'])
output=read.readtext(path)
output
```

**OUTPUT**

```python
top_left=tuple(output[0][0][0])
bottom_right=tuple(output[0][0][2])
text=output[0][1]
font = cv2.FONT_HERSHEY_SIMPLEX
```

```python
img = cv2.imread(path)
img = cv2.rectangle(img,top_left,bottom_right,(0,255,0),3)
img = cv2.putText(img,text,top_left, font, 5,(255,255,255),3,cv2.LINE_AA)
plt.imshow(img)
plt.show()
```
